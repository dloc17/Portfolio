```python
data_species.fillna("No Intervention", inplace = True)
data_species.groupby("conservation_status").size()
```

```python
conservation_category = data_species[data_species.conservation_status != "No Intervention"]\
    .groupby(["conservation_status", "category"])["scientific_name"]\
    .count()\
    .unstack()

conservation_category 
```

```python
ax = conservation_category.plot(kind = 'bar', figsize=(8,6), stacked=True)
ax.set_xlabel("Conservation Status")
ax.set_ylabel("Catgory")
```

```python
data_species["is_protected"] = data_species.conservation_status != "No Intervention"
```

```python
category_counts = data_species.groupby(["category", "is_protected"])\
                            .scientific_name.nunique()\
                            . reset_index()\
                            .pivot(columns="is_protected",
                                   index="category",
                                   values='scientific_name')\
                            .reset_index()
category_counts.columns = ['category', 'not protected', 'protected']

category_counts
```

```python
category_counts["percent protected"] = category_counts.protected / (category_counts.protected + category_counts['not protected']) * 100

category_counts
```

```python
from itertools import chain
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

common_Names = data_species[data_species.category == 'Mammal']\
    .common_names\
    .apply(remove_punctuations)\
    .str.split().tolist()

common_Names[:7]
```

```python
clean_Rows = []

for item in common_Names:
    item = list(dict.fromkeys(item))
    clean_Rows.append(item)

clean_Rows[:7]
```

```python
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in clean_Rows))

res[:7]
```

```python
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i, x))

pd.DataFrame(set(words_counted), columns = ['Word', 'Count']).sort_values("Count", ascending=False).head(10)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                         
```

```python
data_species['is_bat'] = data_species.common_names.str.contains(r'\bBat\b', regex = True)

data_species.head(10)
```

```python
data_species[data_species.is_bat]
```

```python
bat_observations = data_observations.merge(data_species[data_species.is_bat])

bat_observations
```

```python
bat_observations.groupby('park_name').observations.sum().reset_index()
```

```python
obs_by_park = bat_observations.groupby(['park_name', 'is_protected']).observations.sum().reset_index()
obs_by_park
```

```python
plt.figure(figsize = (16, 4))
sns.barplot(x = obs_by_park.park_name, y = obs_by_park.observations, hue = obs_by_park.is_protected)
plt.xlabel("National Parks")
plt.ylabel("Number of Observations")
plt.title("Observations of Bats per Week")
plt.show()
```
